## **Parse dates in time series data**

```
df['time'].value_counts()
```

```
def format_time(x):
  if ':' in x:
    if int(x.split(":")[0]) >= 12 and int(x.split(":")[0]) < 18:
      x = "Afternoon"
    elif int(x.split(":")[0]) < 12 and int(x.split(":")[0]) >= 6:
      x = "Morning"
    elif int(x.split(":")[0]) >= 3 and int(x.split(":")[0]) < 6:
      x = "Early morning"
    elif (int(x.split(":")[0]) >= 18 and int(x.split(":")[0]) < 24) or (int(x.split(":")[0])>=0 and int(x.split(":")[0]) <3):
      x = "Night"
  elif 'evening' in x:
    x = 'Evening'
  elif 'overnight' in x or 'night' in x:
    x = 'Night'
  elif 'morning' in x or 'dawn' in x:
    x = 'Morning'
  elif 'afternoon' in x:
    x = 'Afternoon'
  return x
```

## **Using Bitwise operators to filter rows**

```
df.loc[(df['NumWebPurchases']<5) | (df['NumDealsPurchases']<5), 'ActiveOnlineBuyers']='Not Active'
df.loc[(df['NumWebPurchases']>5) | (df['NumDealsPurchases']>5), 'ActiveOnlineBuyers']='Active'
df
```
