## Assignments don't make a copy of an Array Object 

### *** np.array.copy() ***  also called as  *** deep copy ***

```python
students_id_copy = students_id.copy()
print(students_id_copy)
print(students_id)
```

```python
print(students_id_copy == students_id)
```

```python
print(id(students_id_copy), id(students_id))
```

```python
students_id=students_id/10
print(students_id)
print(students_id_copy)
```

```python
print(id(students_id_copy),id(students_id))
```

```python
students_id_v = students_id.view()
print(students_id_v)
print(students_id)
```

```python
students_id_v[0]=100
print(students_id_v)
print(students_id)
```

```python
## np.array.base() to figure out if it is the original array or a view 

print(students_id.base)
print(students_id_v.base)
```
